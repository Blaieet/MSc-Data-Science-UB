```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
leiStandings = [76,67,66,68,102]
flc = pd.DataFrame({'Points':leiStandings,'Season':['2009-10','2010-11','2011-12','2012-13','2013-14'],
                  'Position':['Standing: 5th','Standing: 10th','Standing: 9th','Standing: 6th','Standing: 1st']})
```

```python
flc.name = "a"
```

```python
flc.name
```

```python
line = alt.Chart(flc).mark_line().encode(
    x=alt.X('Season', title='Season'),
    y=alt.Y('Points:Q', scale=alt.Scale(domain=[40, 120]), title='Points per Season'),
)

nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['Season'], empty='none')

selectors = alt.Chart(flc).mark_point().encode(
    x='Season',
    opacity=alt.value(0),
).add_selection(
    nearest
)

points = line.mark_point(filled=True).encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

text = line.mark_text(align='left', dx=7, dy=-25).encode(
    text=alt.condition(nearest, 'Points:Q', alt.value(' '))
)

text2 = line.mark_text(align='right', dx=20, dy=-27).encode(
    text=alt.condition(nearest, 'Position', alt.value(''))
)


chart = alt.layer(line, selectors, points,text2).configure_axis(
    grid=False
).configure_point(
    size=100
).configure_line(
    size=4
).configure_view(
    stroke='transparent').properties(width=700, height=300).interactive()
chart
```

```python
goals =pd.read_csv('GoalsLeague.csv')

topleague = ['Ligue 1', '1. Bundesliga', 'Premier League', 'Eredivisie', 'LIGA BBVA', 'Serie A']

nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['season'], empty='none')

line = alt.Chart(goals).mark_line().encode(
    x=alt.X('season', title='Season'),
    y=alt.Y('match_goals:Q', scale=alt.Scale(domain=[2.2, 3.4]), title='Goals per game'),
    color=alt.Color('league:N', scale=alt.Scale(domain=topleague))
)
selectors = alt.Chart(goals).mark_point().encode(
    x='season',
    opacity=alt.value(0),
).add_selection(
    nearest
)
goal_last = goals[goals['season'] == '2015/2016']

labels = alt.Chart(goal_last).mark_text(align='left', dx=3).encode(
    alt.X('season'),
    alt.Y('match_goals:Q', scale=alt.Scale(domain=[2.2, 3.4])),
    alt.Text('league'),
    alt.Color('league:N', legend=None, scale=alt.Scale(domain=topleague))
)
points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)
text = line.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'match_goals:Q', alt.value(' '), format='.2f')
)
rules = alt.Chart(goals).mark_rule(color='gray').encode(x='season',).transform_filter(nearest)

chart = alt.layer(line, selectors, labels, points, rules, text).properties(width=1199, height=300).interactive()
chart
```
