1- **Twitter EXERCISE:** 

Given the starting user "eloipuertas", retrieve the user name, the date of the accoount creation and the last tweet date and text for each follower. Store them to a database called 'twitter' and into a collection called 'info'. The attributes of the document are: 'screen_name', 'created_at', 'last_tweet_text' and 'last_tweet_created_at'. 


```python
import json
import pymongo
import tweepy
import time

with open('consumer_key.txt', 'r') as f:
    consumer_key =  f.read()
f.closed

with open('consumer_secret.txt', 'r') as f:
    consumer_secret = f.read()
f.closed

with open('access_key.txt', 'r') as f:
    access_key = f.read()
f.closed

with open('access_secret.txt', 'r') as f:
     access_secret = f.read()
f.closed


#Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)


#Do something
USER_NAME = "neoares92"
user = api.get_user(id=USER_NAME)

ids = []
for page in tweepy.Cursor(api.followers_ids, screen_name=USER_NAME).pages():
    print("************** ***************")
    ids.extend(page)
    time.sleep(5)  #This should be 60 to avoid limit rates


try:
    conn=pymongo.MongoClient()
    print ("Connected successfully!!!")
except pymongo.errors.ConnectionFailure as e:
    print ("Could not connect to MongoDB: %s" % e) 

db = conn['twitter']
collection = db['info']

for id1 in ids:
    document={}
    us=api.get_user(id1)
    document['screen_name']=us.screen_name
    try:
        tweet=api.user_timeline(id = id1, count = 1)[0]
        document['created_at']=us.created_at
        document['last_tweet_text']=tweet.text
        document['last_tweet_created_at']=tweet.created_at
        collection.insert_one(document)
    except Exception as e:
        collection.insert_one(document)
        pass
[d for d in collection.find()]
#Needs a bit of time to execute
```

```python
from selenium import webdriver
import re
browser = webdriver.Chrome("chromedriver_win32/chromedriver.exe")
browser.get("https://www.tripadvisor.es/Restaurants")
html=browser.page_source

element=browser.find_elements_by_class_name('_3qLQ-U8m')[1]
element.clear()
s='sant cugat'
element.send_keys(s)
time.sleep(2)
browser.find_element_by_class_name('_1dvyiAq4').click()
time.sleep(2)
try:
    conn=pymongo.MongoClient()
    print ("Connected successfully!!!")
except pymongo.errors.ConnectionFailure as e:
    print ("Could not connect to MongoDB: %s" % e) 

db = conn['TripAdvisor']
collection = db['Restaurants']

rests=browser.find_elements_by_class_name('_15_ydu6b')
cont=0
for rest in rests:
    if(rest.text[0].isdigit()==False): continue
    document={} 
    document['name']=rest.text
    document['link']=rest.get_attribute('href')
    collection.insert_one(document)
    cont=cont+1
    if(cont==10): break
    
[d for d in collection.find()]
```
